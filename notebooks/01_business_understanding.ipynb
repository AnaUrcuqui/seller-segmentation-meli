# 01 · Comprensión del negocio

**Fase 1 de CRISP-DM**

## Planteamiento del problema

El equipo comercial de MercadoLibre necesita identificar sellers con perfiles sólidos y relevancia para el negocio. Sin una segmentación confiable, es imposible diseñar estrategias comerciales dirigidas a escala.

## Objetivo de negocio

Construir un sistema de segmentación de sellers que:
1. Agrupe a los sellers en clusters significativos y accionables.
2. Permita al equipo comercial priorizar contactos y diseñar campañas diferenciadas.
3. Pueda extenderse con una capa de GenAI que proponga estrategias personalizadas por segmento.

## Hipótesis

| # | Hipótesis | Enfoque de validación |
|---|-----------|----------------------|
| H1 | Los sellers difieren significativamente en tamaño de catálogo y distribución de precios | EDA – estadísticos descriptivos, distribuciones |
| H2 | Un grupo reducido de sellers concentra una proporción desproporcionada del volumen | Análisis de Pareto / concentración |
| H3 | La estrategia de precios (premium vs. descuento) correlaciona con la amplitud del catálogo | Matriz de correlación |
| H4 | Existen arquetipos de sellers diferenciados y recuperables mediante clustering | Silhouette score, perfilado de clusters |

## Criterios de éxito

- **Estadístico**: Silhouette ≥ 0.35 en el clustering final.
- **De negocio**: Cada cluster es interpretable y traducible en al menos una acción comercial concreta.
- **Técnico**: El pipeline se ejecuta de extremo a extremo, desde el CSV crudo hasta etiquetas y estrategias, en menos de 10 min en un portátil estándar.

## Preguntas de los stakeholders a responder

1. ¿Cuántos segmentos de vendedores significativos existen?
2. ¿Qué diferencia a un vendedor de alto valor de uno de cola larga?
3. ¿A qué vendedores debe contactar primero el equipo comercial?
4. ¿Cómo se puede incorporar a un nuevo vendedor y asignarle inmediatamente un segmento?
